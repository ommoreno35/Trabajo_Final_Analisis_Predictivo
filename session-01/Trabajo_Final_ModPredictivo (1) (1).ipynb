```python
Visualización exploratoria 
```

```python
Gráfico 1: Distribución por Género 
```

```python
df['Genero'].value_counts().plot(kind='bar', color='green')
plt.title('Distribución por Género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.show()
```

```python
Gráfico 2: Histograma de edades
```

```python
df['Edad'].dropna().plot(kind='hist', bins=20, edgecolor='skyblue')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
Gráfico 3: Boxplot de Hr_uso_movil por género
```

```python
df.boxplot(column='Hr_uso_movil', by='Genero')
plt.title('Uso Móvil por Género')
plt.suptitle('')
plt.xlabel('Género')
plt.ylabel('Hr uso de Móvil')
plt.show()
```
