```python
CREAR VARIABLE OBJETIVO
```

```python
# Cargar el dataset limpio
df_raw = pd.read_csv('datos/data_salud_mental.csv')

# Crear variable objetivo: Alta Afectación Mental
df_raw["Alta_Afectacion_Mental"] = (df_raw["Puntuación_salud_mental"] > 64).astype(int)

df_raw.head(5)
```

```python
EVALUACIÓN DEL MODELO ANTES DE LIMPIAR LOS DATOS
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Usamos solo Edad e Ingreso, eliminando nulos
df_before = df_raw[["Hr_uso_movil", "Hr_uso_TV", "Hr_redes_sociales"]].dropna()
Xb = df_before
yb = df_raw.loc[df_before.index, "Alta_Afectacion_Mental"]

Xb_train, Xb_test, yb_train, yb_test = train_test_split(Xb, yb, test_size=0.3, random_state=42)

model_before = LogisticRegression(max_iter=200)
model_before.fit(Xb_train, yb_train)
yb_pred = model_before.predict(Xb_test)
acc_before = accuracy_score(yb_test, yb_pred)

print("✅ Exactitud antes de la limpieza:", round(acc_before, 4))
```

```python
PREPARACIÓN DEL DATASET: LIMPIEZA, IMPUTACIÓN Y CODIFICACIÓN
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

df_clean = df_raw.copy()

# Codificación de variables categóricas
df_clean = pd.get_dummies(df_clean, columns=['Genero', 'Tipo_ubicación'], drop_first=True)

# Escalado
scaler = MinMaxScaler()
df_clean[['Hr_uso_movil', 'Hr_uso_portatil', 'Hr_uso_Tablet', 'Hr_uso_TV', 'Hr_redes_sociales', 'Hr_uso_gamming', 'Hr_sueño']] = scaler.fit_transform(df_clean[['Hr_uso_movil', 'Hr_uso_portatil', 'Hr_uso_Tablet', 'Hr_uso_TV', 'Hr_redes_sociales', 'Hr_uso_gamming', 'Hr_sueño']])

```

```python
EVALUACIÓN DEL MODELO DESPUES DE LA LIMPIEZA
```

```python
Xc = df_clean.drop(columns=['user_id', 'Puntuación_salud_mental', 'Alta_Afectacion_Mental'])
yc = df_clean['Alta_Afectacion_Mental']

Xc_train, Xc_test, yc_train, yc_test = train_test_split(Xc, yc, test_size=0.3, random_state=42)

model_after = LogisticRegression(max_iter=100)
model_after.fit(Xc_train, yc_train)
yc_pred = model_after.predict(Xc_test)
acc_after = accuracy_score(yc_test, yc_pred)

print("✅ Exactitud después de la limpieza:", round(acc_after, 4))
```
